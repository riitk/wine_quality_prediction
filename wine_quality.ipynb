```python
X = df.drop("quality", axis = 1)
y = df["quality"].apply(lambda x: 1 if x >=7 else 0)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

**LogisticRegression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
regression = LogisticRegression(max_iter=500)
```

```python
regression.fit(X_train, y_train)
```

```python
reg_predictions = regression.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(accuracy_score(y_test, reg_predictions))
print("")
print(confusion_matrix(y_test, reg_predictions))
print("")
print(classification_report(y_test, reg_predictions))
```

**Support Vector**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel="linear")
```

```python
model.fit(X_train_scaled, y_train)
```

```python
s_predictions = model.predict(X_test_scaled)
```

```python
print(accuracy_score(y_test, s_predictions))
print("")
print(confusion_matrix(y_test, s_predictions))
print("")
print(classification_report(y_test, s_predictions))
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier()
```

```python
forest.fit(X_train, y_train)
```

```python
predictions = forest.predict(X_test)
```

```python
print(accuracy_score(y_test, predictions))
print("")
print(confusion_matrix(y_test, predictions))
print("")
print(classification_report(y_test, predictions))
```
